####  Transforming the Data

```python
# Transforming the Non-Numerical columns into Numerical Columns
# For that we need to import LabelEncoder

from sklearn.preprocessing import LabelEncoder
```

```python
for column in data.columns:
    if data[column].dtype==np.number:
        continue
    data[column]=LabelEncoder().fit_transform(data[column])
```

```python
data.head(7)
```

So, we can see that Encoding is Successful

As our Label is 'Attrition', I want it to appear 1st in the dataset.

```python
# Create a New Column

data['Age_Years']=data['Age'] 
```

```python
data=data.drop('Age', axis=1)
```

#### Splitting the Data

```python
# For that we need some Libraries to be loaded

from sklearn.model_selection import train_test_split
```

```python
X = data.iloc[: , 1:data.shape[1]].values
Y = data.iloc[: , 0].values
```

Now we split the data into 75% Training and 25% Testing  

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25, random_state = 1000) 
```

### Modelling the Data

Using Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 1000)
forest.fit(X_train, Y_train)
```

```python
# Now we can check and get the 'Accuracy' Score

forest.score(X_train, Y_train)
```

```python
# Getting the 'Confusion Matrix' and 'Accuracy Score' for the Model on Test Data

from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
cm = confusion_matrix(Y_test, forest.predict(X_test))

```

```python
accuracyScore = accuracy_score(Y_test, forest.predict(X_test))
```

```python
print(cm)
print(accuracyScore)
```

So, we can see there is a clear improvement in the Accuracy Score of 'Attrition'. 

- Earlier we got a result of only 80.77% for 'No'
- But now it is 84.23%
