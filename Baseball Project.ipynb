```python
data.drop('EarnedRunAvg', axis=1, inplace=True)
```

#### Modelling the Data

```python
# Defining X and Y

X = data.drop(['Wins'],axis=1).values
Y = data['Wins'].values
```

```python
print(X)
```

```python
print(Y)
```

Splitting the Dataset in Training and Testing

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size=0.2, random_state=400)
```

### Training the model on Training set

```python
# Using Linear Regression

from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(X_train,Y_train)
```

```python
Y_pred = lr.predict(X_test)
print(Y_pred)
```

```python
lr.predict([[724,5575,1497,300,42,139,383,973,104,641,601,2,8,56,88]])
```

#### Model Evaluation

```python
from sklearn.metrics import r2_score
```

```python
r2_score(Y_test, Y_pred)
```

We are getting a pretty good r2 score.

#### Plotting the Result

```python
plt.figure(figsize=(10,5))
plt.scatter(Y_test,Y_pred)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs. Predicted')
```

```python
# Predicted Values

pred_Y_df=pd.DataFrame({'Actual Value': Y_test, 'Predicted Value': Y_pred, 'Difference': Y_test-Y_pred})
```
