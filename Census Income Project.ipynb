### Filling the values for '?' 

As we have seen above that there are three Columns with missing values. As they are a kind of Categorical variables we can fill 
the missing values with mode.

```python
data = data.replace('[?]', np.NaN , regex=True)
```

```python
data.info()
```

Now here we can see that there are some missing values in some features and they are - Workclass, Occupation, and Native_country.

As they have huge number of missing values we cannot just drop them it may cause some problem and we may not get the true picture of the data.

```python
data.head(50)
```

```python
data.isnull().sum()
```

As we can see that all the columns with the missing value are Categorical in nature.

We cannot use the Mean for imputation. Here we see that which is the value occurred mostly i.e, MODE we are going to use.

```python
for col in ['Workclass', 'Occupation', 'Native_country']:
    data[col].fillna(data[col].mode()[0], inplace=True)
```

```python
data.isnull().sum()
```

```python
data.head(50)
```

Now we don't have any Missing Value in any of the column. 

#### We have to do some Feature Engineering  

As we have seen in the dataset that there are Features with Categorical Data, and there are many data in it which keep repeating indicating that they belong to a special class. So, we are able to reduce it some more.

 Features on which we are going to apply it are:
 - Education
 - Marital Status
 - Income

```python
# Education

data.Education = data.Education.replace(['Preschool','1st-4th','5th-6th','7th-8th','9th','10th','11th','12th'], 'School', 
                                        regex=True)
data.Education = data.Education.replace('HS-grad','HighSchool', regex=True)
data.Education = data.Education.replace(['Assoc-voc','Assoc-acdm ','Prof-school','Some-college'], 'College', regex=True)
data.Education = data.Education.replace('Bachelors','UnderGrad', regex=True)
```

```python
data['Education'].value_counts()
```

```python
# Marital Status

data.Marital_status = data.Marital_status.replace(['Married-civ-spouse','Married-AF-spouse'], 'Married', regex=True)
data.Marital_status = data.Marital_status.replace('Never-married','Unmarried',regex=True)
data.Marital_status = data.Marital_status.replace(['Divorced','Separated','Widowed','Married-spouse-absent'],'Single', 
                                                  regex=True)
```

```python
data['Marital_status'].value_counts()
```

```python
# Income

data.Income = data.Income.replace('<=50K', 0, regex=True)
data.Income = data.Income.replace('>50K', 1, regex=True)
```

### Model Building

Two models we are going to develop here:
1. Logistic Regression, and
2. Random Forest Classifier

```python
X = data.drop(['Income'], axis=1)
y = data['Income']
```

```python
# FEATURE SCALING

from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
data1 = data.copy()
data1 = data1.apply(LabelEncoder().fit_transform)
```

```python
data1.head()
```

```python
ss = StandardScaler().fit(data1.drop('Income', axis=1))
```

```python
X = ss.transform(data1.drop('Income', axis=1))
y = data1['Income']
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25, random_state=100)
```

##### 1.LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

logreg = LogisticRegression()

model = logreg.fit(X_train, y_train)
pred = model.predict(X_test)

print("Accuracy on Train Data : ",(logreg.score(X_train, y_train)))
print("Accuracy on Test Data : ",(logreg.score(X_test, y_test)))
```

##### 2.RANDOM FOREST CLASSIFIER 

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier()

rfc_model = rfc.fit(X_train, y_train)
rfc_pred = rfc_model.predict(X_test)

print("Accuracy on Train Data : ",(rfc.score(X_train, y_train)))
print("Accuracy on Test Data : ",(rfc.score(X_test, y_test)))
```

Here, we have seen that the Accuracy Score have improved in Random Forest Classifier model in comparison to Logistic Regression by 16% for TRAIN set of Data and by 1.6% for TEST set of Data.
